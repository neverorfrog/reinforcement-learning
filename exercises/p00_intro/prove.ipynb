```python
states = np.array([(random.random(),random.random()) for i in range(10)])*10
state1 = torch.from_numpy(states[0])
print(state1)
state2 = torch.from_numpy(states[1])
print(state2)
state3 = torch.from_numpy(states[2])

replay = deque(maxlen=2)
replay.append((state1,1,state2,0))
replay.append((state2,1,state2,1))

s1_batch = torch.cat([s1 for(s1,a,s2,r) in replay])
print(s1_batch)
```
