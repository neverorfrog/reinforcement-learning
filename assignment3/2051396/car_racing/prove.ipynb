```python
import numpy as np
from collections import namedtuple, deque
import torch

class ERB:
    def __init__(self, memory_size=50000, burn_in=10000):
        self.memory_size = memory_size
        self.burn_in = burn_in
        self.experience = namedtuple('experience',field_names=['state', 'action', 'reward', 'done', 'next_state'])
        self.buffer = deque(maxlen=memory_size)

    def sample_batch(self, batch_size=2):
        samples = np.random.choice(len(self.buffer), batch_size, replace=False)
        # Use asterisk operator to unpack deque
        batch = zip(*[self.buffer[i] for i in samples])
        return batch

    def append(self, s_0, a, r, d, s_1):
        self.buffer.append(self.experience(s_0, a, r, d, s_1))

    def burn_in_capacity(self):
        return len(self.buffer) / self.burn_in

    def capacity(self):
        return len(self.buffer) / self.memory_size

    def __iter__(self):
       ''' Returns the Iterator object '''
       return iter(self.buffer)
```

```python
buffer = ERB()
buffer.append(1,1,1,0,0)
buffer.append(1,0,1,0,1)
buffer.append(1,1,0,0,0)

batch = buffer.sample_batch()
states, actions, rewards, dones, next_states = list(batch)
print(states)
print(next_states)
```
